```python
import json
from datetime import datetime as dt

def Product_Check(prod_id, quant):     #function checks if the Prod is available or not using product id
    if prod_id not in Current_data:
        print("Entered product ID is incorrect. Please try again")
        return False
    
    if Current_data[prod_id]['quant'] == 0:
        print( "We're sorry.OUT OF STOCK")
        return False
    
    if Current_data[prod_id]['quant'] < quant:
        print("Stock is less than demand")
        return False
        
    return True


f = open('records.json', 'r')   #open json file to show menu of current inventory items
txt = f.read()
f.close()

Current_data  = json.loads(txt)

print("**********************************************************WELCOME******************************************************")

print(" Product List".center(120))


print("\n{:<13} {:<25} {:<10} {:<15} {:<13} {:<10} {:<10}".format('PRODUCT-ID', 'PRODUCT-NAME', 'PRICE', 'MFG', 'EXP', 'CATEGORY', 'QUANTITY IN STOCK'))

for product in Current_data:
    print("{:<13} {:<25} {:<10} {:<15} {:<14} {:<15} {:<10}".format(product,Current_data[product]['Product'],Current_data[product]['MRP'],Current_data[product]['mfg'],Current_data[product]['exp'],Current_data[product]['category'],Current_data[product]['quant']))
    


filter_permit = input("\n\nPress 'f' if you want to filter products by category or press any other key to continue: ")

if(filter_permit == 'f'):  #filter products by category
    flag = 0
    while(flag == 0):  #prompting user to enter category to filter products until category is valid
        
        filt = input("Enter the category you Like to Purchase: ").lower()
        
        for product in Current_data:
            if Current_data[product]['category'] == filt:
                print("\n{:<13} {:<25} {:<10} {:<15} {:<14} {:<15} {:<10}".format(product,Current_data[product]['Product'],Current_data[product]['MRP'],Current_data[product]['mfg'],Current_data[product]['exp'],Current_data[product]['category'],Current_data[product]['quant']))
                flag=1

        if flag==0:
            print("\nCategory not found. Please try again\n")

user_prods = {}  #shopping list
while(1):
    inp_id = input("Enter Product ID of the product you want to purchase: ")
    quant =  int(input("Enter quantity required by you: "))
    if quant <=0:
        print("Please enter a valid quantity")
        continue
        
    if(Product_Check(inp_id,quant)):   #function to check if the prod id and quantity are appropriate
        user_prods[inp_id] = quant
        choice = input("Please Enter 'y' for countinue or press any other key for checkout: ")
        if choice == 'y':
            continue
    
        else:
            break
            
#-------------------------------------------------------updating inventory-------------------------------------------------------
for product in user_prods:
    Current_data[product]['quant']-= user_prods[product]
    
    
f = open("records.json",'w')
js = json.dumps(Current_data)
f.write(js)
f.close()



#------------------------------------------------------updating sales file------------------------------------------------------


f=open("sales.json",'r')
txt = f.read()
f.close()
date = str(dt.now().strftime("%Y-%m-%d"))

arr=[]
for p in user_prods:
    name = Current_data[p]['Product']
    ID = p
    quant = user_prods[p]
    amount = quant * Current_data[p]['MRP']
    
    arr.append({'name':name, 'id':ID, 'amount':amount, 'quant':quant})

f = open("sales.json",'a')
txt = json.dumps(arr)
f.write(txt)
f.close()



#------------------------------------------------------------------------------------------------------------------------------




print("\n\nYour shoppping cart has the following: ")
print("\n{:<25} {:<15} {:<10} {:<10}".format('Product','Quantity', 'Price', 'Sub-Total'))
total=0
for prod in user_prods:
    print("{:<25} {:<15} {:<10} {:<10}".format(Current_data[prod]['Product'],user_prods[prod],Current_data[prod]['MRP'],Current_data[prod]['MRP']*user_prods[prod]))
    total+=Current_data[prod]['MRP']*user_prods[prod]

    
print("\n\n\n")
print("PROCEEDING TO CHECKOUT...".center(120))
print("\n\n\n")
print("\n{:<25} {:<30} {:<15} {:<10} ".format(' ','BABA GENERAL STORE','GST No.', '25ASD2544XS1'))
print("---------------------------------------------------------BILL---------------------------------------------------------")
print("\n{:<25} {:<25} {:<15} {:<10} {:<10}".format(' ','Product','Quantity', 'Price', 'Sub-Total'))
for prod in user_prods:
    print("{:<25} {:<25} {:<15} {:<10} {:<10}".format(' ',Current_data[prod]['Product'],user_prods[prod],Current_data[prod]['MRP'],Current_data[prod]['MRP']*user_prods[prod]))
    

print("\n\n{:<25} {:<2} {:<13} {:<2}".format(' ',"      TOTAL AMOUNT: Rs.",total,str(dt.now().strftime("%Y-%m-%d, %H:%M"))))
print("\n----------------------------------------------------------------------------------------------------------------------")
```
